# Ряды

## 1. Исследовать ряд на сходимость, используя признак д’Аламбера:

### $$\sum\limits_{n=1}^{+\infty}\frac{n^n}{(n!)^2}$$

Ряд знакоположительный, поэтому применение признака д’Аламбера допустимо.

### $$a_{n+1}=\frac{(n+1)^{n+1}}{((n+1)!)^2}=\frac{(n+1)^n\cdot(n+1)}{(n!)^2\cdot(n+1)^2}=\frac{(n+1)^{n-1}}{(n!)^2}$$

Найдем предел:

### $$\lim_{n\to +\infty}\frac{a_{n+1}}{a_{n}}=\lim_{n\to +\infty}\frac{(n+1)^{n-1}\cdot(n!)^2}{(n!)^2\cdot n^n}=\lim_{n\to +\infty}\frac{(n+1)^{n-1}}{n^n}$$

Так как степень знаменателя больше степени числителя, то данный предел равен $0$.

$0<1$, значит ряд сходится.

### _Ответ:_ ряд сходится

## 2. Исследовать ряд на сходимость, используя радикальный признак Коши:

### $$\sum\limits_{n=1}^{+\infty}\frac{n}{2^n}$$

Ряд знакоположительный, поэтому применение радикального признака Коши допустимо.

Найдем предел:

### $$\lim_{n\to +\infty}\sqrt[n]{a_{n}}=\lim_{n\to +\infty}\sqrt[n]{\frac{n}{2^n}}=$$

### $$=\lim_{n\to +\infty}\frac{\sqrt[n]{n}}{2}=\frac{1}{2}$$

т.к. предел равен $\frac{1}{2}<1$, то ряд сходится.

### _Ответ:_ ряд сходится

## 3. Исследовать ряд на сходимость, используя признак Лейбница:

### $$\sum\limits_{n=1}^{+\infty}\frac{(-1)^n}{n+\ln n}$$

Проверим необходимые условия:

### a)$$\lim_{n\to +\infty}a_{n}=0$$

### $$\lim_{n\to +\infty}\frac{(-1)^n}{n+\ln n}=0$$

### b)$$|a_n|\geq|a_{n+1}|$$

$|a_n|>|a_{n+1}|$, т.к. $|(-1)^n|=1$, а знаменатель $n+1$ члена будет больше знаменателя $n$ члена, для всех $n>0$.

Необходимые условия выполняются, значит ряд сходится.

Определим характер сходимости. Исследуем ряд по модулю:

### $$\sum\limits_{n=1}^{+\infty}\bigl|\frac{(-1)^n}{n+\ln n}\bigr|=\sum\limits_{n=1}^{+\infty}\frac{1}{n+\ln n}$$

Для сравнения возьмем гармонический ряд:

### $$\lim_{n\to +\infty}\frac{\frac{1}{n}}{\frac{1}{n+\ln n}}=\lim_{n\to +\infty}\frac{n+\ln{n}}{n}=1+\lim_{n\to +\infty}\frac{\ln{n}}{n}=1$$

$0<1<\infty$, и, потому как гармонический ряд расходится, то и исходный ряд расходится, а следовательно знакопеременный ряд сходится условно.

### _Ответ:_ знакопеременный ряд сходится условно.

## 4. Исследовать ряд на сходимость, используя признак Раабе:

### $$\sum\limits_{n=1}^{+\infty}\frac{3^n}{2^n}$$

Ряд знакоположительный, поэтому применение признака Раабе допустимо.

Найдем предел:

### $$\lim_{n\to +\infty}n\Bigr(\frac{a_{n}}{a_{n+1}}-1\Bigl)=\lim_{n\to +\infty}n\Bigr(\frac{3^n\cdot2^{n+1}}{2^n\cdot3^{n+1}}-1\Bigl)=\lim_{n\to +\infty}n\Bigr(\frac{2}{3}-1\Bigl)=-\infty$$

т.к. предел равен $-\infty<1$, то ряд расходится.

### _Ответ:_ ряд расходится

## 5. Разложить функцию по Тейлору в единице:

### $$f(x)=\ln(16x^2)$$

Формула разложения имеет вид:

### $$f(x)=\sum\limits_{n=0}^{+\infty}\frac{f^{(n)}(a)}{n!}(x-a)^n$$

Найдём значение функции в точке $a=1$:

### $$f(1)=\ln(16\cdot1^2)=\ln(16)$$

Найдём производные и их значения в точке $a=1$:

### $$f^{(1)}(x)=\frac{1}{16x^2}\cdot32x=2x^{-1}\Rightarrow f^{(1)}(1)=2$$

### $$f^{(2)}(x)=2\cdot(-x^{-2})\Rightarrow f^{(2)}(1)=-2$$

### $$f^{(3)}(x)=2\cdot(2x^{-3})\Rightarrow f^{(3)}(1)=4$$

### $$f^{(4)}(x)=2\cdot(-2\cdot3x^{-4})\Rightarrow f^{(4)}(1)=-12$$

### $$f^{(n)}(x)=2\cdot(-1)^{n+1}\cdot(n-1)!\cdot x^{-n}\Rightarrow f^{(n)}(1)=2\cdot(-1)^{n-1}\cdot(n-1)!$$

В итоге получаем следующее разложение:

### $$f(x)=\ln(16)+\sum\limits_{n=1}^{+\infty}\frac{2\cdot(-1)^{n-1}\cdot(n-1)!}{n!}(x-1)^n=\ln(16)+\sum\limits_{n=1}^{+\infty}\frac{2\cdot(-1)^{n-1}\cdot(n-1)!}{(n-1)!\cdot n}(x-1)^n=$$

### $$=\ln(16)+2\cdot\sum\limits_{n=1}^{+\infty}\frac{(-1)^{n-1}}{n}(x-1)^n$$

```python
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
import numpy as np

y = lambda x: np.log(16*x**2)

def t(x,n):
    result = 0
    for i in np.arange(1, n+1):
        result += ((-1)**(i-1)/i)*(x-1)**i
    return result

x = np.linspace(0.1, 2.5, 100)
ln16 = np.log(16)

#Максимальное количество членов ряда
nmax = 6

fg = plt.figure(figsize=(20, 20), constrained_layout=True)
gs = gridspec.GridSpec(ncols=2, nrows=1+(nmax-1)//2, figure=fg)

for j in np.arange(1, nmax+1):
    fg.add_subplot(gs[(j-1)//2, (j-1)%2])
    plt.plot(x, y(x), label='Исходная функция f(x)=ln(16x^2)')
    plt.plot(x, ln16 + 2*t(x, j), label=f"Разложение по Тейлору в 1, n={j}")
    plt.legend(loc='best')
    plt.grid(True)

plt.show()
```

### _Ответ:_ $f(x)=\ln(16)+2\cdot\sum\limits_{n=1}^{+\infty}\frac{(-1)^{n-1}}{n}(x-1)^n$

## 6*. Дана функция:

### $$f(x)=x^2$$

## a. Разложить функцию в ряд Фурье по косинусам на отрезке $x\in[-\pi;\,\pi]$

Формула для отрезка $x\in[-\pi;\,\pi]$:

### $$f(x)=a_{0}+\sum\limits_{n=1}^{+\infty}(a_n\cos nx+b_n\sin nx)$$

где

### $$a_0=\frac{1}{2\pi}\int\limits_{-\pi}^\pi f(x)\,dx$$

### $$a_n=\frac{1}{\pi}\int\limits_{-\pi}^\pi f(x)\cos nx\,dx$$

### $$b_n=\frac{1}{\pi}\int\limits_{-\pi}^\pi f(x)\sin nx\,dx$$

Найдем $a_0$:

### $$a_0=\frac{1}{2\pi}\int\limits_{-\pi}^\pi f(x)\,dx=\frac{1}{2\pi}\int\limits_{-\pi}^\pi x^2\,dx=$$

### $$=\frac{1}{2\pi}\cdot\frac{x^3}{3}\Bigr|_{-\pi}^\pi=\frac{1}{2\pi}\cdot\Bigr(\frac{\pi^3}{3}-\frac{(-\pi)^3}{3}\Bigl)=\frac{\pi^2}{3}$$

Найдем $a_n$:

### $$a_n=\frac{1}{\pi}\int\limits_{-\pi}^\pi f(x)\cos nx\,dx=\frac{1}{\pi}\int\limits_{-\pi}^\pi x^2\cos nx\,dx=$$

Данный интеграл будем брать по частям:

### $$U=x^2\,\,\,\,\,\Rightarrow\,\,\,\,\, dU=2x\,dx$$ 
### $$dV=\cos nx\,dx\,\,\,\,\,\Rightarrow\,\,\,\,\, V=\frac{1}{n}\sin nx$$ 

### $$=\Bigr(\frac{1}{\pi n}x^2\sin nx\Bigl)\Bigr|_{-\pi}^\pi-\frac{2}{\pi n}\int\limits_{-\pi}^\pi x\sin nx\,dx=$$

Скобка равно нулю, т.к. $\sin \pi n=\sin (-\pi n)=0$. Интеграл возьмем еще раз по частям:

### $$U=x\,\,\,\,\,\Rightarrow\,\,\,\,\, dU=dx$$ 
### $$dV=\sin nx\,dx\,\,\,\,\,\Rightarrow\,\,\,\,\, V=-\frac{1}{n}\cos nx$$ 

### $$=\Bigr(\frac{2}{\pi n^2}x\cos nx\Bigl)\Bigr|_{-\pi}^\pi-\frac{2}{\pi n^2}\int\limits_{-\pi}^\pi \cos nx\,dx=$$

### $$=\frac{2}{\pi n^2}\pi\cos\pi n-\frac{2}{\pi n^2}(-\pi)\cos\pi n=4\frac{(-1)^n}{n^2}$$

Найдем $b_n$:

### $$b_n=\frac{1}{\pi}\int\limits_{-\pi}^\pi f(x)\sin nx\,dx=\frac{1}{\pi}\int\limits_{-\pi}^\pi x^2\sin nx\,dx=$$

Данный интеграл будем брать по частям:

### $$U=x^2\,\,\,\,\,\Rightarrow\,\,\,\,\, dU=2x\,dx$$ 
### $$dV=\sin nx\,dx\,\,\,\,\,\Rightarrow\,\,\,\,\, V=-\frac{1}{n}\cos nx$$ 

### $$=\Bigr(-\frac{1}{\pi n}x^2\cos nx\Bigl)\Bigr|_{-\pi}^\pi+\frac{2}{\pi n}\int\limits_{-\pi}^\pi x\cos nx\,dx=$$

Скобка равно нулю, т.к. $(\pi)^2\cos \pi n=(-\pi)^2\cos (-\pi n)$. Интеграл возьмем еще раз по частям:

### $$U=x\,\,\,\,\,\Rightarrow\,\,\,\,\, dU=dx$$ 
### $$dV=\cos nx\,dx\,\,\,\,\,\Rightarrow\,\,\,\,\, V=\frac{1}{n}\sin nx$$ 

### $$=\Bigr(\frac{2}{\pi n^2}x\sin nx\Bigl)\Bigr|_{-\pi}^\pi-\frac{2}{\pi n^2}\int\limits_{-\pi}^\pi \sin nx\,dx=$$

### $$=\Bigr(\frac{2}{\pi n^3}\cos nx\Bigl)\Bigr|_{-\pi}^\pi=0$$

Подставив найденные значения в формулу получим следующее разложение:

### $$x^2=\frac{\pi^2}{3}+4\sum\limits_{n=1}^{+\infty}\frac{(-1)^n}{n^2}\cos nx$$

на отрезке $[-\pi, \pi]$

### _Ответ:_ $\frac{\pi^2}{3}+4\sum\limits_{n=1}^{+\infty}\frac{(-1)^n}{n^2}\cos nx$ на отрезке $[-\pi, \pi]$

## b. Построить график функции и ее разложения.

```python
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
import numpy as np

y = lambda x: x**2

def t(x,n):
    result = 0
    for i in np.arange(1, n+1):
        result += ((-1)**i/i**2)*np.cos(i*x)
    return result

x = np.linspace(-np.pi, np.pi, 128)
a0 = np.pi**2/3
#Максимальное количество членов ряда
nmax = 8

fg = plt.figure(figsize=(10, 15), constrained_layout=True)
gs = gridspec.GridSpec(ncols=2, nrows=1+(nmax-1)//2, figure=fg)

for j in np.arange(1, nmax+1):
    fg.add_subplot(gs[(j-1)//2, (j-1)%2])
    plt.plot(x, y(x), label='Исходная функция f(x)=x^2')
    plt.plot(x, a0 + 4*t(x, j), label=f"Разложение Фурье на отрезке [-pi; pi], n={j}")
    plt.legend(loc='best')
    plt.grid(True)

plt.show()
```
