## 1. Найти область определения функции

### $$z=\sqrt{1-x^3}+\ln\bigl(y^2-1\bigr)$$

Функция $z(x,y)$ определена при:

### $$\begin{cases}
   1-x^3\geq0, x\in\mathbb{R},\\
   y^2-1>0, y\in\mathbb{R}
 \end{cases}$$

### $$1-x^3\geq0\Rightarrow x^3\leq1\Rightarrow x\leq1$$

### $$y^2-1>0\Rightarrow y^2>1\Rightarrow y>1\,||\,y<-1$$

### _Ответ:_ $(x,y)\in\mathbb{R}^2:(x\leq1, y<-1)\,||\,(x\leq1, y>1)$

## 2. Найти производные 1-го порядка функции

### $$z'_{y}=3(1+\frac{\ln{x}}{\ln{y}}\bigr)^2\cdot(1+\frac{\ln{x}}{\ln{y}}\bigr)'_{y}=3(1+\frac{\ln{x}}{\ln{y}}\bigr)^2\cdot\bigl(-\frac{\ln{x}}{y\ln^2{y}}\bigr)=$$

### $$=3(\frac{\ln{y}+\ln{x}}{\ln{y}}\bigr)^2\cdot\bigl(-\frac{\ln{x}}{y\ln^2{y}}\bigr)=-\frac{3\ln{x}(\ln{y}+\ln{x})^2}{y\ln^4{y}}$$

### _Ответ:_ $z'_{x}=\frac{3(\ln{y}+\ln{x})^2}{x\ln^3{y}};\,z'_{y}=-\frac{3\ln{x}(\ln{y}+\ln{x})^2}{y\ln^4{y}}$

## 3. Найти полный дифференциал функции в точке (1;1)

### $$z=\sqrt{2xy+\cos{\frac{x}{y}}}$$

Дифференциал функции $z(x,y)$:

### $$dz=\frac{\partial z}{\partial x}dx+\frac{\partial z}{\partial y}dy=z'_{x}dx+z'_{y}dy$$

### $$z'_{x}=\frac{1}{2\sqrt{2xy+\cos{\frac{x}{y}}}}\cdot\bigl(2xy+\cos{\frac{x}{y}}\bigr)'_x=\frac{2y-\frac{\sin{\frac{x}{y}}}{y}}{2\sqrt{2xy+\cos{\frac{x}{y}}}}$$

### $$z'_{y}=\frac{1}{2\sqrt{2xy+\cos{\frac{x}{y}}}}\cdot\bigl(2xy+\cos{\frac{x}{y}}\bigr)'_y=\frac{2x+\frac{x\sin{\frac{x}{y}}}{y^2}}{2\sqrt{2xy+\cos{\frac{x}{y}}}}=\frac{x\bigl(2+\frac{\sin{\frac{x}{y}}}{y^2}\bigr)}{2\sqrt{2xy+\cos{\frac{x}{y}}}}$$

### $$dz=\frac{2y-\frac{\sin{\frac{x}{y}}}{y}}{2\sqrt{2xy+\cos{\frac{x}{y}}}}dx+\frac{x\bigl(2+\frac{\sin{\frac{x}{y}}}{y^2}\bigr)}{2\sqrt{2xy+\cos{\frac{x}{y}}}}dy$$

### $$dz|_{(1,1)}=\frac{2-\sin{1}}{2\sqrt{2+\cos{1}}}dx+\frac{2+\sin{1}}{2\sqrt{2+\cos{1}}}dy$$

### _Ответ:_ $dz|_{(1,1)}=\frac{2-\sin{1}}{2\sqrt{2+\cos{1}}}dx+\frac{2+\sin{1}}{2\sqrt{2+\cos{1}}}dy$

## 4. Исследовать на экстремум функцию

### $$z=x^2+xy+y^2-6x-9y$$

__Необходимым условием__ существования экстремума является __равенство нулю частных производных.__

Найдем частные производные, приравняем их к нулю и найдем стационарные точки:

### $$\begin{cases}
   z'_{x}=2x+y-6=0, \\
   z'_{y}=x+2y-9=0 
 \end{cases}$$

### $$\begin{cases}
   y=6-2x, \\
   x+2(6-2x)-9=0
 \end{cases}\Rightarrow x+12-4x-9=0\Rightarrow x=1,\,y=4$$

Получаем единственную стационарную точку $(1,4)$.

__В качестве основного достаточного условия возьмем критерий Сильвестра.__ Найдем вторые производные и составим матрицу Гёссе:

###  $$z''_{xx}=2$$ $$z''_{xy}=z''_{yx}=1$$  $$z''_{yy}=2$$

###   $$\begin{pmatrix}
2 & 1\\ 
1 & 2\\
\end{pmatrix}$$

Главные миноры 

### $$\Delta_{1}=z''_{xx}=2>0$$

### $$\Delta_{2}=\begin{vmatrix}
z''_{xx} & z''_{xy}\\ 
z''_{yx} & z''_{yy}\\
\end{vmatrix}=z''_{xx}\cdot z''_{yy}-(z''_{xy})^2=2 \cdot 2 - 1^2=3>0$$

Оба минора положительные, это говорит о том, что найденная точка $(1,4)$ является точкой минимума.

```python
import pylab
from mpl_toolkits.mplot3d import Axes3D
import numpy

def makeData ():
    x = numpy.arange (0, 2, 0.1)
    y = numpy.arange (2, 6, 0.1)
    xgrid, ygrid = numpy.meshgrid(x, y)

    zgrid = xgrid**2+xgrid*ygrid+ygrid**2-6*xgrid-9*ygrid
    return xgrid, ygrid, zgrid

x, y, z = makeData()

fig = pylab.figure()
axes = Axes3D(fig)

axes.plot_surface(x, y, z)

pylab.show()
```

### _Ответ:_ $(1, 4)$ - точка минимума функции
